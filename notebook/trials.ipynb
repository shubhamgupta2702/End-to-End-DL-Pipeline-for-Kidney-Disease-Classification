### Ensure Annotations usage and trial

@ensure_annotations is a decorator that used on the top of the function to ensure the typechecking of the input given to the funtion.
no need to explicitly writing the manual typechecking

```python
def multiply(a:int, b:int):
  return a*b
```

```python
multiply(4,"5")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def multiply(a:int, b:int):
  return a*b
```

```python
multiply(4,"5")
```
